## **Categorical Variables and One Hot Encoding**

```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
dum_town = pd.get_dummies(df.town)
dum_town = dum_town.replace({True: 1, False: 0})
dum_town
```

```python
merged = pd.concat([df, dum_town], axis='columns')
merged
```

```python
merged.drop(['town', 'west windsor'], axis='columns', inplace=True)
merged
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = merged.drop('price', axis='columns')
x
```

```python
y = merged.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,1]])
```

```python
model.score(x,y)
```

First plot data points on scatterplot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions :

**2. Predict price of a BMW X5 that is 7 yr old with mileage 86000**

**3. Tell me the score (Accuracy) of model. (Hint: Use LinearRegression.score())**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
car_model = pd.get_dummies(df['Car Model'])
car_model = car_model.replace({True: 1, False: 0})
car_model
```

```python
merged = pd.concat([df, car_model], axis='columns')
merged
```

```python
x = merged.drop(['Sell Price($)',  'Car Model', 'Audi A5'], axis = 1)
x
```

```python
y = merged['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

**Price of mercedez benz that is 4 yr old with mileage 45000**

```python
model.predict([[45000, 4, 0, 1]])
```

**Price of BMW X5 that is 7 yr old with mileage 86000**

```python
model.predict([[86000, 7, 1, 0]])
```

**Accuracy of model**

```python
model.score(x,y)
```
