We have a value of x and y vectors and we want to derive a best fit line or an equation using m and b....

```python
import numpy as np

def gradient_decent (x,y):
  m_curr = b_curr = 0
  iterations = 100
  n = len(x)
  learning_rate = 0.08
  cost_history = [] # Initialize a list to store cost values

  for i in range (iterations):
    y_predicted = m_curr * x + b_curr
    cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
    cost_history.append(cost) # Append the cost for this iteration
    md = -(2/n)*sum(x*(y-y_predicted))
    bd = -(2/n)*sum(y-y_predicted)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd
    print ("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost, i))

  return m_curr, b_curr, cost_history # Return cost history along with m and b

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Call the function and store the results, including the cost history
m, b, cost_history = gradient_decent(x, y)

print("Final values: m =", m, ", b =", b)
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(12, 5))

# Plot the cost function over iterations
plt.subplot(1, 2, 1)
plt.plot(cost_history) # Use the returned cost_history
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function over Iterations')

# Plot the regression line and data points
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', label='Data Points')
y_predicted = m * x + b
plt.plot(x, y_predicted, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()

plt.tight_layout()
plt.show()
```
